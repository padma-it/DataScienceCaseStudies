#### 3.2.4 OBSERVATION: Get the columns in which percentage of null values > 40% and  Plot the columns vs Null Values Present in it

```python
null_applicationDF = pd.DataFrame((applicationDF.isnull().sum())*100/applicationDF.shape[0]).reset_index()
null_applicationDF.columns = ['Column Name', 'Null Values Percentage']
null_applicationDF
```

```python
# Plotting the missing values
fig = plt.figure(figsize=(18,8))
ax = sns.pointplot(x="Column Name",y="Null Values Percentage",data=null_applicationDF)
plt.xticks(rotation =90,fontsize =7)
ax.axhline(40, ls='--',color='red')
plt.title("Percentage of Missing values in application data")
plt.ylabel("Null Values PERCENTAGE")
plt.xlabel("COLUMNS")
plt.show()
```

#### 3.2.5: Repeat of (3.1.4) FOR DATA CLEANING: Get the columns in which percentage of null values > 40%

```python
cols = applicationDF.columns[applicationDF.isnull().mean()>=0.4]
applicationDF.drop(cols, axis=1,inplace=True)
applicationDF.columns
```
