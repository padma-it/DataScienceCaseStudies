## 3. Data Cleaning



### 3.1 Null Values Investigation
#### 3.1.1 Get the number of null values in each column

```python
#Checking the null values for each columns
applicationDF.isnull().sum()
```

#### 3.1.2 Get the percentage of null values in each column

```python
#Checking the null value % for each columns
round(applicationDF.isnull().sum() / applicationDF.shape[0] * 100.00,2)
```

#### 3.1.3 Get the number of rows before data cleaning

```python
original_total_rows_count = len(applicationDF.index)
original_total_rows_count
```

#### 3.1.4 OBSERVATION: Get the columns in which percentage of null values > 40% and  Plot the columns vs Null Values Present in it

```python
null_applicationDF = pd.DataFrame((applicationDF.isnull().sum())*100/applicationDF.shape[0]).reset_index()
null_applicationDF.columns = ['Column Name', 'Null Values Percentage']
null_applicationDF
```

```python
# Plotting the missing values
fig = plt.figure(figsize=(18,8))
ax = sns.pointplot(x="Column Name",y="Null Values Percentage",data=null_applicationDF)
plt.xticks(rotation =90,fontsize =7)
ax.axhline(40, ls='--',color='red')
plt.title("Percentage of Missing values in application data")
plt.ylabel("Null Values PERCENTAGE")
plt.xlabel("COLUMNS")
plt.show()
```

#### 3.1.5: Repeat of (3.1.4) FOR DATA CLEANING: Get the columns in which percentage of null values > 40%

```python
# Get more than or equal to 40% empty rows columns into a dataframe
null_col_40 = null_applicationDF[null_applicationDF['Null Values Percentage']>=40]
null_col_40.columns = ['Column Name', 'Null Values Percentage']
null_col_40
```

```python
# create a list of columns which has more than or equal to 40% null values as use it to drop them from main dataframe : 
Unwanted = null_col_40["Column Name"].tolist()
Unwanted
```

```python
# how many columns are getting deleted
len(Unwanted)
```

```python
applicationDF.drop(labels=Unwanted,axis=1,inplace=True)
```

```python
applicationDF.columns
```

```python
applicationDF.isnull().sum(axis=1) 
```
