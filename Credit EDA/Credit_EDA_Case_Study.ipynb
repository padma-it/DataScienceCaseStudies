## 3. Data Cleaning



#### 3.1 Get the number of null values in each column

```python
#Checking the null values for each columns
applicationDF.isnull().sum()
```

#### 3.2 Get the percentage of null values in each column

```python
#Checking the null value % for each columns
round(applicationDF.isnull().sum() / applicationDF.shape[0] * 100.00,2)
```

#### 3.3 Get the number of rows before data cleaning

```python
original_total_rows_count = len(applicationDF.index)
original_total_rows_count
```

#### 3.4 Get the columns in which percentage of null values > 50%

```python
round(applicationDF.isnull().sum() / original_total_rows_count * 100.00,2) > 50
```

```python
applicationDF.isnull().sum(axis=1) 
```

#### 3.5 Drop unnecessary columns

```python
drop_column_names = ['','']
applicationDF.drop(columns=drop_column_names, axis=1, inplace=True)
```

```python
# Write your code for checking number of retained rows here
# Get the new column-wise null percentages and observe the differences
new_total_rows_count = len(applicationDF.index)
print("Columns which still have null values:{0}\nAll columns Overview:\n{1} ".format(applicationDF.columns[round(((applicationDF.isnull().sum(axis=0) / new_total_rows_count) * 100),2) > 0], round(((applicationDF.isnull().sum(axis=0) / new_total_rows_count) * 100),2)))

# Get new total number of rows
print("\nPercentage of rows retained: ", round(((new_total_rows_count / original_total_rows_count ) * 100),2))
```
