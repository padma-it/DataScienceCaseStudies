- ### Subtask 3.5: Find the best directors

    1. Group the dataframe using the `director_name` column.
    2. Find out the top 10 directors for whom the mean of `imdb_score` is the highest and store them in a new dataframe `top10director`.  Incase of a tie in IMDb score between two directors, sort them alphabetically. 

```python
# Extract the top 10 directors here using steps:
# Create a groupby object and apply mean function; 
# rename the mean applied column; 
# sort the resultant dataframe; 
# Get only top 10 directors and reset the index
top10director = movies.groupby('director_name').aggregate({'imdb_score':np.mean}).rename(columns={'imdb_score':'Mean IMDb score'}).sort_values(['Mean IMDb score', 'director_name'], ascending=[False,True]).iloc[0:10].reset_index()

top10director
```

-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Extract the first two genres of each movie here
genre_df = movies['genres'].str.split("|", expand = True)
movies['genre_1'] = genre_df.loc[:,0] # Add genre1 column
movies['genre_2'] = genre_df.loc[:,1] # Add genre2 column

single_genred_movies = movies['genre_2'].isnull()
print("Total number of movies with a single genre: ", len(movies.loc[single_genred_movies]) )
movies.loc[single_genred_movies,'genre_2'] = movies['genre_1'] #  Assign genre 1 to genre 2 for single-genred movies
print("Total number of movies with a single genre: ", len(movies.loc[movies['genre_2'].isnull()]) )
```

```python
# Group the dataframe here by genres
movies_by_segment = movies.groupby(['genre_1', 'genre_2']) 
```

```python
# Get the 5 most popular combo of genres here
PopGenre = movies_by_segment['gross'].mean().sort_values(ascending=False).iloc[0:5].reset_index()
PopGenre
```

**Checkpoint 5:** Well, as it turns out. `Family + Sci-Fi` is the most popular combo of genres out there!

    1. Create three new dataframes namely, `Meryl_Streep`, `Leo_Caprio`, and `Brad_Pitt` which contain the movies in which the actors: 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' are the lead actors. Use only the `actor_1_name` column for extraction. Also, make sure that you use the names 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' for the said extraction.
    2. Append the rows of all these dataframes and store them in a new dataframe named `Combined`.
    3. Group the combined dataframe using the `actor_1_name` column.
    4. Find the mean of the `num_critic_for_reviews` and `num_users_for_review` and identify the actors which have the highest mean.
    5. Observe the change in number of voted users over decades using a bar chart. Create a column called `decade` which represents the decade to which every movie belongs to. For example, the  `title_year`  year 1923, 1925 should be stored as 1920s. Sort the dataframe based on the column `decade`, group it by `decade` and find the sum of users voted in each decade. Store this in a new data frame called `df_by_decade`.

```python
# Create three new dataframes here

Meryl_Streep = movies.loc[movies['actor_1_name']=='Meryl Streep'] # Include all movies in which Meryl_Streep is the lead
Meryl_Streep 
```

```python
Leo_Caprio = movies.loc[movies['actor_1_name']=='Leonardo DiCaprio'] # Include all movies in which Leo_Caprio is the lead
Leo_Caprio
```

```python
Brad_Pitt = movies.loc[movies['actor_1_name']=='Brad Pitt'] # Include all movies in which Brad_Pitt is the lead
Brad_Pitt
```

```python
# Combine the three dataframes here
Combined = Meryl_Streep.append([Leo_Caprio,Brad_Pitt])
Combined
```

```python
# Group the combined dataframe here
df_by_segment = Combined.groupby('actor_1_name')
df_by_segment
```

```python
# Find the mean of critic reviews and audience reviews here using the steps:

# Create a groupby object and apply mean functions on two columns; 
# rename the mean applied columns; 
# sort the resultant dataframe; 
# Observe Leonardo is the top on both lists

Combined_DF_Actor_Grouped = df_by_segment.aggregate({'num_critic_for_reviews':np.mean,'num_user_for_reviews':np.mean}).rename(columns={'num_critic_for_reviews':'Mean num_critic_for_reviews', 'num_user_for_reviews':'Mean num_user_for_reviews'}).sort_values(['Mean num_critic_for_reviews','Mean num_user_for_reviews'],ascending=False)

Combined_DF_Actor_Grouped
```

**Checkpoint 6:** `Leonardo` has aced both the lists!

```python
# Calculate decade here using steps:
# calculate decade by converting it to int; 1923
# and then divide by 10 will give modulo value (1923%10 -> 3)
# subtract that from the actual year (1923.0-3 = 1920.0)
# convert to int to trim out the decimals and then convert to string and add 's' at the end
movies['decade'] = (movies['title_year'] - (movies['title_year'].astype(int))%10).astype(int).astype(str) + 's'
movies.sort_values(by='decade', inplace=True)
movies
```

```python
# Creating the data frame df_by_decade here using steps:
# Create groupby and rename the column 
df_by_decade = pd.DataFrame(movies.groupby('decade')["num_user_for_reviews"].sum()).rename(columns={'num_user_for_reviews':'Total num_user_for_reviews'}).reset_index()
df_by_decade
```

```python
# Plot number of voted users vs decade using both matplotlib and seaborn(log scale)
plt.figure(figsize=(15,15))
plt.subplot(121)
plt.bar(x=df_by_decade['decade'], height=df_by_decade['Total num_user_for_reviews'],color = "steelblue", edgecolor = "orange")

plt.style.use('Solarize_Light2')
plt.xlabel("Decade",fontdict={'fontsize' : 10, 'fontweight' : 3, 'color' : 'Blue'})
plt.ylabel("Total number of user reviews",fontdict={'fontsize' : 10, 'fontweight' : 3, 'color' : 'Blue'})
plt.title("Total number of user reviews trend over decades", fontdict={'fontsize' : 15, 'fontweight' : 5, 'color' : 'Blue'}) # styling with dictionary

tick_values = np.arange(0,700000,50000)
tick_labels = ["0L","5L", "10L","15L", "20L","25L", "30L", "35L","40L", "45L", "50L", "55L","60L","65L"]
plt.yticks(tick_values, tick_labels) # Added granular yticks for better visibility of bars
# Add values in the bars - position of label(x,y) is at (i-0.3),(df_by_decade['num_user_for_reviews'][i]+10000) for better visibility
for i in range(10):
    plt.text(i-0.3,df_by_decade['Total num_user_for_reviews'][i]+10000,int(df_by_decade['Total num_user_for_reviews'][i]))

# Write your code for plotting number of voted users vs decade - using seaborn and log scale for more visibility of bars
plt.subplot(122)
plt.yscale('log')
sns.barplot(x='decade', y='Total num_user_for_reviews', data=df_by_decade)
plt.title("Total number of user reviews trend over decades")
for i in range(10):
    plt.text(i-0.4,df_by_decade['Total num_user_for_reviews'][i]+1000,int(df_by_decade['Total num_user_for_reviews'][i]))
plt.show()
```
